## Data Preprocessing

```python
#Natural language processing libraries are imported
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
clean_review_list = []
review_len_list = []
```

```python
def preprocess(raw_tweet):
    # Remove Special Character
    letters_only = re.sub('[^a-zA-Z]',' ',raw_tweet)
    
    # Conver sentence into Lower Case
    letters_only = letters_only.lower()
    
    # Tokenize
    words = letters_only.split()
    
    #Remove Stop Words
    words = [w for w in words if not w in stopwords.words('english')]
    
    # Stemming
    
   
    words = [stemmer.stem(word) for word in words]
    
        
    clean_review_list.append(' '.join(words))  
    review_len_list.append(len(words))
```

```python
from tqdm import tqdm
tqdm.pandas()
```

```python
amzcomb_df['ReviewText'].progress_apply(lambda x: preprocess(x))

amzcomb_df['Clean_review_stem'] = clean_review_list
amzcomb_df['review_len'] = review_len_list
```

## Model Building and Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
classifiers = [
                LogisticRegression(),
                DecisionTreeClassifier(),
                KNeighborsClassifier(n_neighbors = 8),
                svm.SVC()
]
```

## Prediction on Future Data Points

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
def preprocess_test(raw_review):
    # Remove Special Character
    letters_only = re.sub('[^a-zA-Z]',' ',raw_review)
    
    # Conver sentence into Lower Case
    letters_only = letters_only.lower()
    
    # Tokenize
    words = letters_only.split()
    
    #Remove Stop Words
    words = [w for w in words if not w in stopwords.words('english')]
    
    # Stemming
    
    words = [stemmer.stem(word) for word in words]
    
    clean_review = ' '.join(words)
    
    return clean_review 
```

```python
def predict(review):
    
    vectorizer = load(open('vect.pkl','rb'))
    model = load(open('amzfood_logistic.pkl','rb'))
    
    clean_review = preprocess_test(review)
    
    clean_review_vector = vectorizer.transform([clean_review])
    
    prediction = model.predict(clean_review_vector)
    
    return prediction
```
